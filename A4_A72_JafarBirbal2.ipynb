Aim:
To perform data ingestion and cleaning using Pandas.

**Theory:**
Data ingestion is the process of collecting and importing data from various sources into a system for analysis. These sources can include databases, CSV files, APIs, web scraping, or real-time streaming data. The main goal of data ingestion is to gather raw data and store it in a structured format so it can be easily accessed and analyzed. In Python, libraries like Pandas are commonly used to load data using functions such as read_csv(), read_excel(), or read_json().


```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
df.to_csv("titanic.csv", index=False)
df = pd.read_csv("titanic.csv")
df.head()
df.info()
df.head()
df.tail()
df.isnull().sum()
# Fill missing Age with average
df["Age"] = df["Age"].fillna(df["Age"].mean())
# Drop Cabin column (too many missing)
df = df.drop("Cabin", axis=1)
df = df.drop_duplicates()
df.isnull().sum()
df.head()
```

Q4. Difference between head() and tail()

The head() function in Pandas displays the first five rows of a DataFrame by default, which is useful for a quick initial inspection of the data. On the other hand, the tail() function displays the last five rows of a DataFrame by default, which is useful for checking the end of the dataset, especially after operations that might append data.

Both functions are excellent for quickly getting a sense of your data's structure and content without printing the entire DataFrame.
